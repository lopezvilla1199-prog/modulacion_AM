```python
# ============================================
# SIMULACIÓN DE MODULACIÓN AM EN PYTHON
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# ============================================
# 1. DEFINICIÓN DE LA SEÑAL DE ENTRADA
# ============================================

fs = 1000
t = np.arange(0, 1, 1/fs)

fm = 5
fc = 50
Am = 1
Ac = 2

mensaje = Am * np.sin(2*np.pi*fm*t)

# ============================================
# 2. SEÑAL PORTADORA
# ============================================

portadora = Ac * np.sin(2*np.pi*fc*t)

# ============================================
# 3. MODULACIÓN AM
# ============================================

senal_modulada = (1 + mensaje) * portadora

# ============================================
# 4. GRÁFICAS EN EL TIEMPO
# ============================================

plt.figure(figsize=(10,6))

plt.subplot(3,1,1)
plt.plot(t, mensaje)
plt.title("Señal mensaje")

plt.subplot(3,1,2)
plt.plot(t, portadora)
plt.title("Señal portadora")

plt.subplot(3,1,3)
plt.plot(t, senal_modulada)
plt.title("Señal AM modulada")

plt.tight_layout()
plt.show()

# ============================================
# 5. ANÁLISIS EN FRECUENCIA
# ============================================

N = len(t)
yf = fft(senal_modulada)
xf = fftfreq(N, 1/fs)

plt.plot(xf[:N//2], np.abs(yf[:N//2]))
plt.title("Espectro de frecuencia señal modulada")
plt.xlabel("Frecuencia")
plt.ylabel("Magnitud")
plt.show()

# ============================================
# 6. RUIDO
# ============================================

ruido = np.random.normal(0, 0.5, len(t))
senal_ruidosa = senal_modulada + ruido

plt.plot(t, senal_ruidosa)
plt.title("Señal modulada con ruido")
plt.show()

# ============================================
# 7. ATENUACIÓN
# ============================================

senal_atenuada = 0.5 * senal_modulada

plt.plot(t, senal_atenuada)
plt.title("Señal modulada con atenuación")
plt.show()

# ============================================
# 8. DISTORSIÓN
# ============================================

senal_distorsionada = np.clip(senal_modulada, -1, 1)

plt.plot(t, senal_distorsionada)
plt.title("Señal modulada con distorsión")
plt.show()
```
